```python
import os
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Path to the folder containing the CSV files
folder_path = "C:/Users/rmcarri/OneDrive - Emory University/NHANES/Macronutrients LLMs/02.Datasets"

# List to store individual DataFrames
df_list = []

# Loop over all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        df_list.append(df)

# Concatenate all DataFrames into one
combined_df = pd.concat(df_list, ignore_index = True)

# Optional: display the result
print(combined_df.shape)
combined_df.head()
```

```python
# Create new ID that combines SEQN and Survey Year

combined_df['id'] = combined_df['SEQN'].astype(str) + "_" + combined_df['id_study']
combined_df
```

```python
"""
NHANES Protocol:
Interviews were conducted with a proxy for participants less than six years of age (who was generally the person most knowledgeable about the participant's intake).
Interviews of children aged 6 to 8 were conducted with a proxy and the child was present to assist in reporting intake information.
Interviews of children aged 9-11, were conducted with the child and the assistance of a proxy familiar with the child's intake.
Participants 12 years or older answered for themselves. Dietary interviewers conducted telephone interviews in English and Spanish.
"""

combined_df = combined_df[(combined_df['age'] >= 12) & (combined_df['age'] <= 19)]
print(combined_df['age'].describe())
print('==========' *10)
combined_df[['age', 'id_study']].groupby('id_study').count()
```

```python
combined_df.describe()
```

```python
# Create a mask for the condition - Men cannot be pregnant
mask1 = (combined_df['is_preg'].isna()) & (combined_df['sex'] == 1)
combined_df.loc[mask1, 'is_preg'] = 0

# Create a mask for the condition - As per NHANES protocol, pregnancy asked for women 20-44 years
mask2 = (combined_df['is_preg'].isna()) & (combined_df['sex'] == 2)
combined_df.loc[mask2, 'is_preg'] = 0


print(combined_df['is_preg'].value_counts(dropna=False))
print('==========' *10)
print(pd.crosstab(combined_df['sex'], combined_df['is_preg'], dropna=False))
```

```python
filtered_df_10_examples = combined_df.sample(n = 10, random_state = 42)
print(filtered_df_10_examples.shape)
filtered_df_10_examples.to_csv('C:/Users/rmcarri/OneDrive - Emory University/NHANES/Macronutrients LLMs/Combined_df_10_examples.csv', index=False)
```

```python
filtered_df_remaining = combined_df.drop(filtered_df_10_examples.index)


print(filtered_df_remaining.shape)
filtered_df_remaining.to_csv('C:/Users/rmcarri/OneDrive - Emory University/NHANES/Macronutrients LLMs/Combined_df_ten_shot.csv', index=False)


filtered_df_remaining[filtered_df_remaining['day'] == 1].to_csv('C:/Users/rmcarri/OneDrive - Emory University/NHANES/Macronutrients LLMs/Combined_df_ten_shot_day1.csv', index=False)
filtered_df_remaining[filtered_df_remaining['day'] == 2].to_csv('C:/Users/rmcarri/OneDrive - Emory University/NHANES/Macronutrients LLMs/Combined_df_ten_shot_day2.csv', index=False)
```

```python
print(f"Total Sample Size: {filtered_df_remaining.shape[0]}")
print(f"Sample Size Day 1: {filtered_df_remaining[filtered_df_remaining['day'] == 1].shape[0]}")
print(f"Sample Size Day 2: {filtered_df_remaining[filtered_df_remaining['day'] == 2].shape[0]}")
print(f"Sample in Day 1 + Sample Day 2 = Total Sample?: {(filtered_df_remaining[filtered_df_remaining['day'] == 1].shape[0] + filtered_df_remaining[filtered_df_remaining['day'] == 2].shape[0]) == filtered_df_remaining.shape[0]}")
print(f"Number of duplicated IDs in day 1: {filtered_df_remaining[filtered_df_remaining['day'] == 1]['id'].duplicated().sum()}")
print(f"Number of duplicated IDs in day 2: {filtered_df_remaining[filtered_df_remaining['day'] == 2]['id'].duplicated().sum()}")
```

```python
# Split day 2 in chunks, so that we can run each as an indipendent job and save time

filtered_df_day2 = filtered_df_remaining[filtered_df_remaining['day'] == 2]   # Get the filtered DataFrame
filtered_df_day2 = filtered_df_day2.reset_index(drop = True)
chunks = np.array_split(filtered_df_day2, 10)   # Split into 10 chunks
for i, chunk in enumerate(chunks, 1):           # Save each chunk with the specified filename pattern
    print(f"Sample size in chunk {i}: {chunk.shape[0]}")
    chunk.to_csv(
        f'C:/Users/rmcarri/OneDrive - Emory University/NHANES/Macronutrients LLMs/Combined_df_ten_shot_day2_chunk{i}.csv',
        index = False
    )
```
